# Traida de datos originales

```python
import pandas as pd

# Rutas a los archivos CSV
archivo_product = 'CSV/amazon_product.csv'
archivo_review = 'CSV/amazon_review.csv'

# Leer los archivos CSV en DataFrames
df_productos = pd.read_csv(archivo_product) 
df_review = pd.read_csv(archivo_review)

# Mostrar los encabezados de cada DataFrame
print("Encabezados de df_productos:")
print(df_productos.columns.tolist())

print("\nEncabezados de df_review:")
print(df_review.columns.tolist())

# Contar el número de filas en cada DataFrame
print("\nNúmero de datos en df_productos:", df_productos.shape[0])
print("Número de datos en df_review:", df_review.shape[0])

# Contar los nulos en cada columna de df_productos
print("\nNúmero de nulos por columna en df_productos:")
print(df_productos.isnull().sum())

# Contar los nulos en cada columna de df_review
print("\nNúmero de nulos por columna en df_review:")
print(df_review.isnull().sum())

# Contar el número total de filas duplicadas en df_productos
print("\nNúmero total de filas duplicadas en df_productos:", df_productos.duplicated().sum())

# Contar duplicados por columna en df_productos
print("\nNúmero de duplicados por columna en df_productos:")
for col in df_productos.columns:
    print(f"{col}: {df_productos.duplicated(subset=[col]).sum()}")

# Contar el número total de filas duplicadas en df_review
print("\nNúmero total de filas duplicadas en df_review:", df_review.duplicated().sum())

# Contar duplicados por columna en df_review
print("\nNúmero de duplicados por columna en df_review:")
for col in df_review.columns:
    print(f"{col}: {df_review.duplicated(subset=[col]).sum()}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df_product = pd.read_csv('CSV/amazon_product_clean.csv')

# Eliminar el símbolo ₹ y convertir a número
df_product['discounted_price'] = df_product['discounted_price'].replace({'₹': ''}, regex=True)
df_product['discounted_price'] = pd.to_numeric(df_product['discounted_price'], errors='coerce')

# Convertir a entero, manejando NaNs
df_product['discounted_price'] = df_product['discounted_price'].fillna(0).astype(int)

sns.set(style="whitegrid", rc={"axes.facecolor": "#f0f0f0", "grid.color": "#704214"})  # Fondo gris claro y grid sepia

plt.figure(figsize=(12, 6))
sns.boxplot(x=df_product['discounted_price'], color='#ADD8E6')
plt.title('discounted_price')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from IPython.display import HTML

# Leer el archivo CSV en un DataFrame
df_product = pd.read_csv('CSV/amazon_product_clean.csv')

# Ordenar el DataFrame por 'discount_percentage' de mayor a menor
df_product_sorted = df_product.sort_values(by='discount_percentage', ascending=False)

# Función para ajustar el formato de las celdas en HTML
def render_dataframe(df, max_rows=10):
    html = df.head(max_rows).to_html(index=False, escape=False)
    # Ajustar el estilo de las celdas para truncar texto y limitar el ancho
    html = html.replace('<td>', '<td style="max-width: 200px; overflow: hidden; text-overflow: ellipsis; white-space: nowrap;">')
    return html

# Mostrar los primeros 10 registros con el ajuste de truncamiento
display(HTML(render_dataframe(df_product_sorted)))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_product = pd.read_csv('CSV/amazon_product_clean.csv')

# Convertir 'discount_percentage' a números y multiplicar por 100
df_product['discount_percentage'] = pd.to_numeric(df_product['discount_percentage'], errors='coerce')
df_product['discount_percentage'] = df_product['discount_percentage'] * 100

sns.set(style="whitegrid", rc={"axes.facecolor": "#f0f0f0", "grid.color": "#704214"})  # Fondo gris claro y grid sepia

plt.figure(figsize=(12, 6))
sns.boxplot(x=df_product['discount_percentage'], color='#ADD8E6')
plt.title('Discount Percentage')

# Configurar los ticks del eje x para que muestren valores en porcentaje
max_value = df_product['discount_percentage'].max()
plt.xticks(ticks=range(0, int(max_value) + 10, 10), labels=[f'{i}%' for i in range(0, int(max_value) + 10, 10)], rotation=45)

plt.tight_layout()
plt.show()
```
