```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import HTML
from sqlalchemy import create_engine
from IPython.display import display
# Crear una cadena de conexi√≥n para SQLAlchemy
engine = create_engine('postgresql+psycopg2://osiris:121041@localhost:5432/datalab-osiris')

# Consulta para obtener el encabezado de la tabla amazon_review_clean
query_review_clean = "SELECT * FROM amazon_unificado"
amazon_unificado = pd.read_sql_query(query_review_clean, engine)

print(amazon_unificado.columns.tolist())

print("\nEncabezado del Unificado")
# Usa display() para mostrar el DataFrame
#display(amazon_unificado)  
```

```python
from collections import Counter
import re
import pandas as pd

# Concatenar 'review_title' y 'review_content'
amazon_unificado['title_content'] = amazon_unificado['review_title'] + " " + amazon_unificado['review_content']

# Funci√≥n para limpiar y contar palabras
def contar_palabras(texto):
    # Convertir a min√∫sculas
    texto = texto.lower()
    # Eliminar caracteres especiales y dividir en palabras
    palabras = re.findall(r'\b\w{4,}\b', texto)
    # Contar palabras
    return Counter(palabras)

# Aplicar la funci√≥n a toda la columna 'title_content'
conteo_palabras = amazon_unificado['title_content'].apply(contar_palabras)

# Sumar todos los contadores para obtener las 25 palabras m√°s comunes
suma_conteos = sum(conteo_palabras, Counter())
palabras_comunes = [palabra for palabra, _ in suma_conteos.most_common(25)]

# Crear columnas para cada una de las 25 palabras m√°s comunes
for palabra in palabras_comunes:
    amazon_unificado[palabra] = amazon_unificado['title_content'].apply(lambda x: x.lower().split().count(palabra))

# Seleccionar solo las columnas de inter√©s
amazon_resultado = amazon_unificado[['title_content'] + palabras_comunes]

# Exportar el DataFrame resultante a un archivo HTML
amazon_resultado.to_html('amazon_resultado.html', index=False)

# Convertir el conteo total a DataFrame para un mejor display
df_suma_conteos = pd.DataFrame(suma_conteos.items(), columns=['Palabra', 'Conteo']).sort_values(by='Conteo', ascending=False)

# Exportar el DataFrame de conteo total a un archivo HTML
df_suma_conteos.to_html('conteo_total.html', index=False)

print("Los archivos se han exportado como 'amazon_resultado.html' y 'conteo_total.html'.")
```

```python
import pandas as pd
import numpy as np
from IPython.display import display

# DataFrame para almacenar datos num√©ricos
amazon_numericos = pd.DataFrame()

# Recorrer cada columna y validar tipo de dato
for column in amazon_unificado.columns:
    try:
        # Verificar si la columna tiene valores num√©ricos
        if amazon_unificado[column].apply(lambda x: str(x).replace('.', '', 1).isdigit()).all():
            # Convertir a tipo num√©rico
            amazon_unificado[column] = pd.to_numeric(amazon_unificado[column], errors='coerce')
            amazon_numericos[column] = amazon_unificado[column]
        else:
            # Si es string, mantener en el DataFrame original
            amazon_unificado[column] = amazon_unificado[column].astype(str)
    except Exception as e:
        print(f"Error procesando la columna '{column}': {e}")

# Manejo de valores nulos - Llenar nulos con la mediana para evitar NaN
amazon_numericos.fillna(amazon_numericos.median(), inplace=True)

# Mensaje si el DataFrame de datos num√©ricos est√° vac√≠o
if amazon_numericos.empty:
    print("No se encontraron columnas num√©ricas en el DataFrame.")

# Medidas de tendencia central y dispersi√≥n para datos num√©ricos, agrupadas por 'general_category'
try:
    if not amazon_numericos.empty:
        # Agregar la columna 'general_category' al DataFrame num√©rico
        amazon_numericos['general_category'] = amazon_unificado['general_category']
        
        # Agrupar por 'general_category' y calcular las estad√≠sticas b√°sicas
        tendencia_central = amazon_numericos.groupby('general_category').agg(
            Media=('discounted_price', 'mean'),
            Desviaci√≥n_Est√°ndar=('discounted_price', 'std'),
            Varianza=('discounted_price', 'var'),
            Mediana=('discounted_price', 'median'),
            Primer_Cuartil_25=('discounted_price', lambda x: np.percentile(x.dropna(), 25)),
            Tercer_Cuartil_75=('discounted_price', lambda x: np.percentile(x.dropna(), 75)),
            M√≠nimo=('discounted_price', 'min'),
            M√°ximo=('discounted_price', 'max'),
            Moda=('discounted_price', lambda x: x.mode()[0] if not x.mode().empty else np.nan)
        )
        
        # Calcular el rango intercuartil (IQR)
        tendencia_central['Rango_Intercuartil'] = tendencia_central['Tercer_Cuartil_75'] - tendencia_central['Primer_Cuartil_25']

        # Transponer la tabla para que las categor√≠as est√©n en filas y las medidas en columnas
        tendencia_central = tendencia_central.T
        
        # Mostrar la tabla resultante
        display(tendencia_central)
    else:
        print("No se pueden calcular medidas de tendencia central porque no hay columnas num√©ricas.")
except Exception as e:
    print(f"Error calculando las medidas de tendencia central: {e}")

# Verificaci√≥n de errores en columnas string
for column in amazon_unificado.columns:
    if amazon_unificado[column].dtype == 'object':
        if any(amazon_unificado[column].str.contains('[^0-9.]', regex=True)):
            print(f"Advertencia: La columna '{column}' contiene caracteres no num√©ricos.")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurarse de que las columnas 'rating' y 'rating_count' sean num√©ricas
amazon_unificado['rating'] = pd.to_numeric(amazon_unificado['rating'], errors='coerce')
amazon_unificado['rating_count'] = pd.to_numeric(amazon_unificado['rating_count'], errors='coerce')

# Paso 1: Identificaci√≥n del Tipo de Variables
rating_is_numeric = pd.api.types.is_numeric_dtype(amazon_unificado['rating'])
rating_count_is_integer = pd.api.types.is_integer_dtype(amazon_unificado['rating_count'])

if rating_is_numeric and rating_count_is_integer:
    variable_type = "Rating es continuo y Rating Count es entero."
    print(f"Variables: 'rating' es continuo y num√©rico, y 'rating_count' es entero.")

    # Paso 2: Evaluaci√≥n de la Distribuci√≥n de los Datos
    print("\nPaso 2: Evaluaci√≥n de la Distribuci√≥n de los Datos")

    # Test de normalidad para 'rating'
    rating_shapiro = stats.shapiro(amazon_unificado['rating'].dropna())
    print(f"Shapiro-Wilk Test para 'rating': p-valor = {rating_shapiro.pvalue:.4f}")

    # Test de normalidad para 'rating_count'
    rating_count_shapiro = stats.shapiro(amazon_unificado['rating_count'].dropna())
    print(f"Shapiro-Wilk Test para 'rating_count': p-valor = {rating_count_shapiro.pvalue:.4f}")

    # Visualizaci√≥n de la distribuci√≥n
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.histplot(amazon_unificado['rating'].dropna(), kde=True, color='blue')
    plt.title('Distribuci√≥n de Rating')

    plt.subplot(1, 2, 2)
    sns.histplot(amazon_unificado['rating_count'].dropna(), kde=True, color='green')
    plt.title('Distribuci√≥n de Rating Count')
    plt.show()

    # Paso 3: Selecci√≥n del Test Estad√≠stico
    print("\nPaso 3: Selecci√≥n del Test Estad√≠stico")
    if rating_shapiro.pvalue > 0.05 and rating_count_shapiro.pvalue > 0.05:
        # Ambas variables siguen una distribuci√≥n normal
        print("Ambas variables siguen una distribuci√≥n normal. Se utilizar√° el Test de Correlaci√≥n de Pearson y el Test de Correlaci√≥n de Kendall.")
        corr_test1 = "pearson"
        correlation1, p_value1 = stats.pearsonr(amazon_unificado['rating'].dropna(), amazon_unificado['rating_count'].dropna())

        # Test adicional: Kendall para comparaci√≥n
        corr_test2 = "kendall"
        correlation2, p_value2 = stats.kendalltau(amazon_unificado['rating'].dropna(), amazon_unificado['rating_count'].dropna())
        
    else:
        # Al menos una variable no sigue una distribuci√≥n normal
        print("Al menos una variable no sigue una distribuci√≥n normal. Se utilizar√° el Test de Correlaci√≥n de Spearman y el Test de Correlaci√≥n de Kendall.")
        corr_test1 = "spearman"
        correlation1, p_value1 = stats.spearmanr(amazon_unificado['rating'].dropna(), amazon_unificado['rating_count'].dropna())

        # Test adicional: Kendall para comparaci√≥n
        corr_test2 = "kendall"
        correlation2, p_value2 = stats.kendalltau(amazon_unificado['rating'].dropna(), amazon_unificado['rating_count'].dropna())

    # Paso 4: Verificaci√≥n de Supuestos y Resultados
    print("\nPaso 4: Verificaci√≥n de Supuestos y Resultados")

    def describe_correlation(correlation):
        if correlation == 1:
            return "perfecta y positiva"
        elif correlation == -1:
            return "perfecta y negativa"
        elif 0.7 < correlation < 1:
            return "fuerte y positiva"
        elif -1 < correlation < -0.7:
            return "fuerte y negativa"
        elif 0.4 < correlation <= 0.7:
            return "moderada y positiva"
        elif -0.7 <= correlation < -0.4:
            return "moderada y negativa"
        elif 0.2 < correlation <= 0.4:
            return "d√©bil y positiva"
        elif -0.4 <= correlation < -0.2:
            return "d√©bil y negativa"
        else:
            return "muy d√©bil"

    # Resultados del Test 1
    print(f"Test 1 Seleccionado: {corr_test1.capitalize()} Correlation Test")
    print(f"Coeficiente de Correlaci√≥n: {correlation1:.4f} ({describe_correlation(correlation1)})")
    print(f"p-valor: {p_value1:.4f}")
    if p_value1 < 0.05:
        print(f"Existe una correlaci√≥n estad√≠sticamente significativa entre 'rating' y 'rating_count' seg√∫n el test de {corr_test1.capitalize()}.")
    else:
        print(f"No se encontr√≥ una correlaci√≥n estad√≠sticamente significativa entre 'rating' y 'rating_count' seg√∫n el test de {corr_test1.capitalize()}.")

    # Resultados del Test 2
    print(f"\nTest 2 Seleccionado: {corr_test2.capitalize()} Correlation Test")
    print(f"Coeficiente de Correlaci√≥n: {correlation2:.4f} ({describe_correlation(correlation2)})")
    print(f"p-valor: {p_value2:.4f}")
    if p_value2 < 0.05:
        print(f"Existe una correlaci√≥n estad√≠sticamente significativa entre 'rating' y 'rating_count' seg√∫n el test de {corr_test2.capitalize()}.")
    else:
        print(f"No se encontr√≥ una correlaci√≥n estad√≠sticamente significativa entre 'rating' y 'rating_count' seg√∫n el test de {corr_test2.capitalize()}.")

    # Gr√°fico de dispersi√≥n con l√≠nea de tendencia
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='rating', y='rating_count', data=amazon_unificado, color='purple', s=50)
    sns.regplot(x='rating', y='rating_count', data=amazon_unificado, scatter=False, color='red')
    plt.title('Relaci√≥n entre Rating y Rating Count')
    plt.xlabel('Rating')
    plt.ylabel('Rating Count')
    plt.show()

    # Gr√°fico de caja para Rating Count seg√∫n Rating
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='rating', y='rating_count', data=amazon_unificado, hue='rating', palette='coolwarm', legend=False)
    plt.title('Distribuci√≥n de Rating Count por Rating')
    plt.xlabel('Rating')
    plt.ylabel('Rating Count')
    plt.show()

else:
    print("Una o ambas variables no son num√©ricas. No se puede realizar la correlaci√≥n.")
```

```python
import pandas as pd
from IPython.display import display

# Supongamos que tu DataFrame se llama amazon_unificado y ya est√° cargado

# Aseg√∫rate de que discounted_price sea num√©rico
amazon_unificado['discounted_price'] = pd.to_numeric(amazon_unificado['discounted_price'], errors='coerce')

# Calcular los cuartiles para discounted_price
quartiles = amazon_unificado['discounted_price'].quantile([0.25, 0.50, 0.75]).values

# Crear etiquetas para los cuartiles
labels = [
    f'Menor a {quartiles[0]:.2f}',
    f'Menor a {quartiles[1]:.2f}',
    f'Menor a {quartiles[2]:.2f}',
    f'Mayores de {quartiles[2]:.2f}'
]

# Crear una lista para almacenar los DataFrames de cada cuartil
resultados_cuartiles = []

# Funci√≥n para aplicar formato a la raz√≥n
def highlight_reason(val):
    if isinstance(val, (int, float)) and val > 1:
        return 'background-color: lightcoral; color: white; font-weight: bold'
    return ''

# Calcular los resultados para cada cuartil
for i, label in enumerate(labels):
    # Determinar el rango para el cuartil actual
    if i == 0:
        df_cuartil = amazon_unificado[amazon_unificado['discounted_price'] <= quartiles[0]]
    elif i == 1:
        df_cuartil = amazon_unificado[(amazon_unificado['discounted_price'] > quartiles[0]) & (amazon_unificado['discounted_price'] <= quartiles[1])]
    elif i == 2:
        df_cuartil = amazon_unificado[(amazon_unificado['discounted_price'] > quartiles[1]) & (amazon_unificado['discounted_price'] <= quartiles[2])]
    else:
        df_cuartil = amazon_unificado[amazon_unificado['discounted_price'] > quartiles[2]]
    
    # Contar los valores de category_rating
    conteo_alto = df_cuartil[df_cuartil['category_rating'] == 'Alto'].shape[0]
    conteo_bajo = df_cuartil[df_cuartil['category_rating'] == 'Bajo'].shape[0]
    
    # Calcular el total y la proporci√≥n de "Bajo"
    total = conteo_alto + conteo_bajo
    proporcion_bajo = conteo_bajo / total if total > 0 else 0
    
    # Calcular el conteo global de otros cuartiles
    df_otros = amazon_unificado[~amazon_unificado.index.isin(df_cuartil.index)]
    conteo_alto_otros = df_otros[df_otros['category_rating'] == 'Alto'].shape[0]
    conteo_bajo_otros = df_otros[df_otros['category_rating'] == 'Bajo'].shape[0]
    total_otros = conteo_alto_otros + conteo_bajo_otros
    proporcion_bajo_otros = conteo_bajo_otros / total_otros if total_otros > 0 else 0
    
    # Calcular la raz√≥n
    razon_proporcion = proporcion_bajo / proporcion_bajo_otros if proporcion_bajo_otros > 0 else 0
    
    # Crear un DataFrame para el cuartil actual
    resultados_cuartil = pd.DataFrame({
        'Rango de Precio': [label],
        'Conteo Alto': [conteo_alto],
        'Conteo Bajo': [conteo_bajo],
        'Total': [total],
        'Proporci√≥n Bajo': [proporcion_bajo]
    })
    
    # Crear un DataFrame para los dem√°s cuartiles combinados
    resultados_otros = pd.DataFrame({
        'Rango de Precio': ['Todos los dem√°s cuartiles'],
        'Conteo Alto': [conteo_alto_otros],
        'Conteo Bajo': [conteo_bajo_otros],
        'Total': [total_otros],
        'Proporci√≥n Bajo': [proporcion_bajo_otros]
    })
    
    # Crear un DataFrame para la raz√≥n
    resultados_razon = pd.DataFrame({
        'Rango de Precio': ['Raz√≥n'],
        'Conteo Alto': [''],
        'Conteo Bajo': [''],
        'Total': [''],
        'Proporci√≥n Bajo': [razon_proporcion]
    })
    
    # Concatenar los DataFrames del cuartil, los dem√°s cuartiles y la raz√≥n
    resultado_completo = pd.concat([resultados_cuartil, resultados_otros, resultados_razon], ignore_index=True)
    
    # Aplicar el estilo condicional usando map
    styled_resultado = resultado_completo.style.applymap(highlight_reason, subset=['Proporci√≥n Bajo'])
    
    # Agregar los DataFrames a la lista de resultados de cuartiles
    resultados_cuartiles.append(styled_resultado)

# Mostrar los resultados de cada cuartil en Jupyter Notebook
for styled_df in resultados_cuartiles:
    display(styled_df)
```

```python
import pandas as pd

# Supongamos que amazon_unificado es tu DataFrame

# Funci√≥n para determinar el tipo de dato de cada columna
def infer_column_types(df):
    types = {}
    for column in df.columns:
        # Obtener los datos √∫nicos de la columna
        unique_values = df[column].dropna().unique()
        
        # Determinar el tipo de dato
        if pd.api.types.is_numeric_dtype(df[column]):
            types[column] = 'Num√©rico'
        elif pd.api.types.is_string_dtype(df[column]):
            if all(value.replace('.', '', 1).isdigit() for value in unique_values if isinstance(value, str)):
                types[column] = 'Num√©rico (como texto)'
            else:
                types[column] = 'Texto'
        elif pd.api.types.is_datetime64_any_dtype(df[column]):
            types[column] = 'Fecha/Hora'
        else:
            types[column] = 'Desconocido'
    
    return pd.DataFrame(list(types.items()), columns=['Columna', 'Tipo de Dato'])

# Crear la tabla con el tipo de dato de cada columna
column_types_df = infer_column_types(amazon_unificado)

# Mostrar la tabla
print(column_types_df)
```

### Kruskal-Wallis H

```python
import pandas as pd
import numpy as np
from scipy import stats
from IPython.display import display

# Cargar el DataFrame df_unificado
df_unificado = amazon_unificado

# Definir las combinaciones espec√≠ficas
specific_combinations = [
    ('Electronics', 'RemoteControls'),
    ('Computers&Accessories', 'USBCables'),
]

# Resultado esperado como DataFrame
results = []

# Iterar sobre las combinaciones espec√≠ficas
for general_cat, specific_cat in specific_combinations:
    # Filtrar datos por categor√≠a general y espec√≠fica
    specific_subset = df_unificado[
        (df_unificado['general_category'] == general_cat) &
        (df_unificado['specific_category'] == specific_cat)
    ]
    
    # Convertir a valores num√©ricos, ignorando errores y eliminando NaN
    actual_prices = pd.to_numeric(specific_subset['actual_price'], errors='coerce').dropna()
    
    # Verificar si hay suficientes datos para realizar el test
    if len(actual_prices) < 3:
        results.append([
            general_cat,
            specific_cat,
            'No aplicable (menos de 3 datos)',
            len(actual_prices),
            np.nan,
            'No se puede realizar el test de normalidad debido a la falta de datos',
            'No se puede realizar el test de normalidad'
        ])
        continue
    
    # Realizar el test de normalidad de Shapiro-Wilk
    _, p_value_shapiro = stats.shapiro(actual_prices)
    is_normal = p_value_shapiro > 0.05
    
    # Definir el grupo de comparaci√≥n
    comparison_subset = df_unificado[
        (df_unificado['general_category'] == general_cat) &
        (df_unificado['specific_category'] != specific_cat)
    ]
    
    # Convertir a valores num√©ricos, ignorando errores y eliminando NaN
    comparison_prices = pd.to_numeric(comparison_subset['actual_price'], errors='coerce').dropna()
    
    if len(comparison_prices) < 3:
        p_value = np.nan
        result_message = 'No se puede realizar el test de significancia debido a la falta de datos en el grupo de comparaci√≥n'
    else:
        # Realizar el test adecuado
        if is_normal:
            # Test t para muestras independientes si la distribuci√≥n es normal
            _, p_value = stats.ttest_ind(actual_prices, comparison_prices, equal_var=False)
        else:
            # Test Mann-Whitney U si la distribuci√≥n no es normal
            _, p_value = stats.mannwhitneyu(actual_prices, comparison_prices)
        result_message = "El valor p es calculado y registrado."
    
    # Tama√±o de la muestra
    sample_size = len(actual_prices)
    
    # Hip√≥tesis redactada
    hypothesis = (
        f"No hay una diferencia significativa en el precio actual de los productos entre la categor√≠a "
        f"espec√≠fica '{specific_cat}' y otras categor√≠as bajo '{general_cat}'" if p_value > 0.05 else
        f"Hay una diferencia significativa en el precio actual de los productos entre la categor√≠a "
        f"espec√≠fica '{specific_cat}' y otras categor√≠as bajo '{general_cat}'"
    )
    
    # Resultado de la hip√≥tesis
    result = "No se rechaza la hip√≥tesis nula" if p_value > 0.05 else "Se rechaza la hip√≥tesis nula"
    
    # A√±adir los resultados a la lista
    results.append([
        general_cat,
        specific_cat,
        'Normal' if is_normal else 'No normal',
        sample_size,
        p_value,
        hypothesis,
        result
    ])

# Crear un DataFrame con los resultados
results_df = pd.DataFrame(results, columns=[
    'Categor√≠a General', 'Categor√≠a Espec√≠fica', 'Tipo de Distribuci√≥n',
    'Tama√±o de la Muestra', 'p_value', 'Hip√≥tesis', 'Resultado'
])

# Mostrar los resultados usando display
display(results_df)
```
